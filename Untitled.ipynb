```python
import os
patients = []
labels = []
i=0
skip = 3
for filename in os.listdir("resized/"):
    if not filename.startswith("Pat3Train"):
        continue
    #if(i > 10):
    #    break
    if filename[-5] == '0' and skip > 0:
        skip -= 1
        continue
    elif skip < 1 :
        skip = 3
        
    i+=1
    data = np.load("resized/"+filename)
    patients.append(data)
    labels.append(filename[-5] == '1')
```

```python
label_array = np.reshape(np.array(labels, dtype=bool),(-1,1))
pat_arr = np.array(patients)
idx = np.array(range(0,pat_arr.shape[0]))
np.random.shuffle(idx)
pat_arr = pat_arr[idx]
label_array = label_array[idx]
labels_filter = label_array.ravel().copy()
print(labels_filter[0:100])

print(pat_arr.shape, label_array.shape)
```

```python
#pat1s = pat_arr[labels_filter]
#pat0s = pat_arr[~labels_filter]
#pat_arr = np.vstack((pat1s,pat0s[:pat1s.shape[0]]))

test_size = 300



train = np.array(range(pat_arr.shape[0]-test_size))
test = np.array(range(pat_arr.shape[0]-test_size, pat_arr.shape[0]))

train1_labels = label_array[train][labels_filter[train]]
train0_labels = label_array[train][~labels_filter[train]]
test1_labels = label_array[test][labels_filter[test]]
test0_labels = label_array[test][~labels_filter[test]]


train1_pat = pat_arr[train][labels_filter[train]]
train0_pat = pat_arr[train][~labels_filter[train]]

test1_pat = pat_arr[test][labels_filter[test]]
test0_pat = pat_arr[test][~labels_filter[test]]

print(train1_pat.shape)
print(test1_pat.shape)
print(train0_pat.shape)
print(test0_pat.shape)
```

```python
#label1s = label_array[labels_filter]
#label0s = label_array[~labels_filter]
#label_array = np.vstack((label1s,label0s[:label1s.shape[0]]))
print(label_array.shape)
```

```python
import keras as k
import keras.layers as l

nb_features = pat_arr.shape[2]
nb_out = label_array.shape[1]
sequence_length = pat_arr.shape[1]

model = k.Sequential()

reg = 1e-4

model.add(l.LSTM(
           input_shape=(sequence_length, nb_features),
           units = 80,
           return_sequences=True,
           kernel_regularizer=k.regularizers.l2(reg),
           activity_regularizer=k.regularizers.l1(reg)))

model.add(l.Dropout(0.4))
model.add(l.LSTM(
           units = 40,
           return_sequences=False,
           kernel_regularizer=k.regularizers.l2(reg),
           activity_regularizer=k.regularizers.l1(reg)))
model.add(l.Dropout(0.7))
model.add(l.Dense(units=nb_out, activation='sigmoid',kernel_regularizer=k.regularizers.l2(reg),
                activity_regularizer=k.regularizers.l1(reg)))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
def f1(x,y):
    return x*y/(x+y)
def roc(inputs, labels, model):
    return sklearn.metrics.roc_auc_score(labels,model.predict_proba(inputs))
```

```python
def model_creator(reg = 1e-6, units_1 = 20, units_2 = 10, dropout = 0.2):
    model = k.Sequential()

    model.add(l.LSTM(
               input_shape=(sequence_length, nb_features),
               units = units_1,
               return_sequences=True,
               kernel_regularizer=k.regularizers.l2(reg),
               activity_regularizer=k.regularizers.l1(reg)))

    model.add(l.Dropout(dropout))
    model.add(l.LSTM(
               units = units_2,
               return_sequences=False,
               kernel_regularizer=k.regularizers.l2(reg),
               activity_regularizer=k.regularizers.l1(reg)))
    model.add(l.Dropout(dropout))
    model.add(l.Dense(units=nb_out, activation='sigmoid',kernel_regularizer=k.regularizers.l2(reg),
                    activity_regularizer=k.regularizers.l1(reg)))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

max_model = 0
max_roc = 0
max_reg = 0
max_units = (0,0)
max_dropout = -1
for reg in [1e-10,1e-8,1e-7,1e-6]:
    for (units_1,units_2) in [(20,10),(40,20),(80,40),(50,50)]:
        for dropout in [0,0.1,0.2,0.3,0.4]:
            new_model = model_creator(reg, units_1, units_2, dropout)
            new_model.fit(pat_arr[train], label_array[train],batch_size=10,epochs=7)
            new_roc = roc(test,new_model)
            print(new_roc)
            if(new_roc > max_roc):
                max_roc = new_roc
                max_model = new_model
                max_reg = reg
                max_units = (units_1,units_2)
                max_dropout = dropout
```
