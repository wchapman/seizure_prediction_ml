```python
import os
patients = []
labels = []
i=0
for filename in os.listdir("data/"):
    if(i > 10000):
        break
    i+=1
    data = utils.load_file("data/"+filename)
    resized = skimage.transform.resize(data, (2400,16), anti_aliasing=True)
    patients.append(resized)
    labels.append(filename[-6] == '1')
```

```python
label_array = np.reshape(np.array(labels),(-1,1))
labels_filter = label_array.ravel().copy()
pat_arr = np.array(patients)
print(pat_arr.shape, label_array.shape)
```

```python
pat1s = pat_arr[labels_filter]
pat0s = pat_arr[~labels_filter]
pat_arr = np.vstack((pat1s,pat0s[:pat1s.shape[0]]))

print(pat_arr.shape)
```

```python
label1s = label_array[labels_filter]
label0s = label_array[~labels_filter]
label_array = np.vstack((label1s,label0s[:label1s.shape[0]]))
print(label_array.shape)
```

```python
import keras as k
import keras.layers as l

nb_features = pat_arr.shape[2]
nb_out = label_array.shape[1]
sequence_length = pat_arr.shape[1]

model = k.Sequential()

model.add(l.LSTM(
           input_shape=(sequence_length, nb_features),
           units = 100,
           return_sequences=True))
model.add(l.LSTM(
           input_shape=(sequence_length, nb_features),
           units = 100,
           return_sequences=True))

model.add(l.Dropout(0.2))
model.add(l.LSTM(
           units = 50,
           return_sequences=False))
model.add(l.Dropout(0.2))
model.add(l.Dense(units=nb_out, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(pat_arr, label_array,batch_size=10,epochs=50)
```
